```python
# Removed unwanted columns 
df.drop(['Unnamed: 15','Unnamed: 16'],axis=1,inplace=True)
df
```

```python
# setting -200 values as null values
df.replace(to_replace=-200,value=np.nan,inplace=True)
```

```python
df.drop(columns=['NMHC(GT)'],inplace=True)
df
```

```python
df.drop(['Date','Time'],axis=1,inplace=True)
```

```python
df.columns
```

```python
print(len(df.columns))
```

```python
df.isna().sum()
```

```python
col=['CO(GT)','PT08.S1(CO)','C6H6(GT)','PT08.S2(NMHC)','NOx(GT)','PT08.S3(NOx)','NO2(GT)','PT08.S4(NO2)','PT08.S5(O3)','T','RH','AH']
```

```python
df=df[col]
df[col].dtypes
```

```python
df[col].head()
```

```python
# replacing all the null values with mean of each column
for i in col:
    df[i]=df[i].fillna(df[i].mean())
df.isna().sum()
```

```python
# using boxplot to get outliers that present in the dataset
sb.boxplot(data=df)
mp.xticks(rotation='vertical')
mp.show()
```

```python
# removing outliers
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).sum()
```

```python
outliers=(df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))
outliers
```

```python
# replacing complete columns outliers using thier median
for i in outliers.columns:
    df[i].astype('float')
    temp=df[i].median()
    df.loc[outliers[i],i]=temp
```

```python
# check for pending outliers
((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).sum()
```

```python
# boxplot to represent outliers
sb.boxplot(data=df)
mp.xticks(rotation='vertical')
mp.show()
```

```python
x=df[['CO(GT)','PT08.S1(CO)','C6H6(GT)','PT08.S2(NMHC)','PT08.S3(NOx)','PT08.S4(NO2)','PT08.S5(O3)', 'T', 'RH',
       'AH']]
```

```python
x
```

```python
y=df['NO2(GT)']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
y_train
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred_lr=model.predict(x_test)
y_pred_lr
```

```python
import joblib
joblib.dump(model,"linear_model.pkl")
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,root_mean_squared_error,r2_score
lr_mae=mean_absolute_error(y_test,y_pred_lr)
lr_mape=mean_absolute_percentage_error(y_test,y_pred_lr)
lr_mse=mean_squared_error(y_test,y_pred_lr)
lr_rmse=root_mean_squared_error(y_test,y_pred_lr)
lr_r2_scr=r2_score(y_test,y_pred_lr)
```

```python
df1=pd.DataFrame({'Linear Regression Method':['Mean Absolute Error','Mean Absolute percentage Error','Mean Squared Error','Root Mean Squared Error','r2 Score'],
                  'Values':[lr_mae,lr_mape,lr_mse,lr_rmse,lr_r2_scr]})
df1.sort_values(by='Values',ascending=True)
df1
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor()
tree.fit(x_train,y_train)
y_pred_dt=tree.predict(x_test)
y_pred_dt
```

```python
joblib.dump(tree,"decision_tree_model.pkl")
```

```python
dt_mae=mean_absolute_error(y_test,y_pred_dt)
dt_mape=mean_absolute_percentage_error(y_test,y_pred_dt)
dt_mse=mean_squared_error(y_test,y_pred_dt)
dt_rmse=root_mean_squared_error(y_test,y_pred_dt)
dt_r2_scr=r2_score(y_test,y_pred_dt)
```

```python
df2=pd.DataFrame({'Decision Tree Method':['Mean Absolute Error','Mean Absolute percentage Error','Mean Squared Error','Root Mean Squared Error','r2 Score'],
                  'Score':[dt_mae,dt_mape,dt_mse,dt_rmse,dt_r2_scr]})
df2.sort_values(by='Score',ascending=True)
df2
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor
random_forest=RandomForestRegressor()
random_forest.fit(x_train,y_train)
y_pred_rf=random_forest.predict(x_test)
y_pred_rf
```

```python
joblib.dump(random_forest,"random_forest_model.pkl")
```

```python
rf_mae=mean_absolute_error(y_test,y_pred_rf)
rf_mape=mean_absolute_percentage_error(y_test,y_pred_rf)
rf_mse=mean_squared_error(y_test,y_pred_rf)
rf_rmse=root_mean_squared_error(y_test,y_pred_rf)
rf_r2_scr=r2_score(y_test,y_pred_rf)
```

```python
df3=pd.DataFrame({'Random Forest Method':['Mean Absolute Error','Mean Absolute percentage Error','Mean Squared Error','Root Mean Squared Error','r2 Score'],
                  'Score':[rf_mae,rf_mape,rf_mse,rf_rmse,rf_r2_scr]})
df3.sort_values(by='Score',ascending=True)
df3
```

```python
# Support Vector Machine
from sklearn.svm import SVR
s_v_m=SVR()
s_v_m.fit(x_train,y_train)
y_pred_svm=s_v_m.predict(x_test)
y_pred_svm

```

```python
joblib.dump(s_v_m,"svr_model.pkl")
```

```python
svm_mae=mean_absolute_error(y_test,y_pred_svm)
svm_mape=mean_absolute_percentage_error(y_test,y_pred_svm)
svm_mse=mean_squared_error(y_test,y_pred_svm)
svm_rmse=root_mean_squared_error(y_test,y_pred_svm)
svm_r2_scr=r2_score(y_test,y_pred_svm)
```

```python
df4=pd.DataFrame({'Support Vector Machine method':['Mean Absolute Error','Mean Absolute percentage Error','Mean Squared Error','Root Mean Squared Error','r2 Score'],
                  'Score':[svm_mae,svm_mape,svm_mse,svm_rmse,svm_r2_scr]})
df4.sort_values(by='Score',ascending=True)
df4
```

```python
def risk_levels(value):
    if value < 50:
        return "LOW RISK"
    elif value < 100:
        return "MODERATE RISK"
    else:
        return "High Risk"
```
